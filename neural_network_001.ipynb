```python
import pandas as pd

df = pd.read_csv("insurance-data/insurance_data.csv")  # Update filename if different
df.head()  # Show first 5 rows
```

till now, we have downloaded and un-zipped our insurance dataset

```python
# do train and test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(df[['age']],df.bought_insurance,test_size=0.2,random_state=25)
```

```python
y_test
```

```python
# prompt: max age in x_train and x_test

print("Max age in x_train:", x_train['age'].max())
print("Max age in x_test:", x_test['age'].max())
```

```python
# scaling the dataset so that age is between 0 and 1 so that all the features
# are on the same scale that is between 0 and 1

x_train_scaled=x_train.copy()
x_train_scaled['age']=x_train_scaled['age']/62

x_test_scaled=x_test.copy()
x_test_scaled['age']=x_test_scaled['age']/62
```

```python
x_train_scaled
```

```python
# build a single neural network using keras and tensorflow

model=keras.Sequential([
    keras.layers.Dense(1,input_shape=(1,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train_scaled,y_train,epochs=5000)
```

```python
model.predict(x_test_scaled)
```
